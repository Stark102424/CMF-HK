# BINOMIAL TREE

```python
# Define the parameters for the binomial tree option pricing
S0 = spot_price_jan_2024  # Spot price from NSRGY data
K = strike_price  # Strike price calculated previously
T = time_to_maturity  # Time to maturity (assumed to be 1 month)
r = interest_rate  # Interest rate (3% annually)
sigma =log_volitality  # Volatility from NSRGY data
steps = 100  # Number of steps in the binomial tree

# Calculate delta T and other required parameters
dt = T / steps
u = np.exp(sigma * np.sqrt(dt))
d = 1 / u
p = (np.exp(r * dt) - d) / (u - d)
disc = np.exp(-r * dt)

# Initialize binomial tree matrices for stock prices and option values
price_tree = np.zeros((steps + 1, steps + 1))
option_tree = np.zeros((steps + 1, steps + 1))

# Populate the stock price tree
for i in range(steps + 1):
    for j in range(i + 1):
        price_tree[j, i] = S0 * (u ** (i - j)) * (d ** j)

# Calculate the option price at maturity (European Put Option)
for j in range(steps + 1):
    option_tree[j, steps] = max(K - price_tree[j, steps], 0)

# Work backwards to get option price at the current time
for i in range(steps - 1, -1, -1):
    for j in range(i + 1):
        option_tree[j, i] = disc * (p * option_tree[j, i + 1] + (1 - p) * option_tree[j + 1, i + 1])
        
        

# Option value at the current time
option_value = option_tree[0, 0]

# Visualize the binomial tree (showing only the first few steps for clarity)
plt.figure(figsize=(10, 6))
for i in range(min(steps, 3) + 1):
    for j in range(i + 1):
        plt.plot(i, price_tree[j, i], 'bo')  # Stock price nodes
        if i < min(steps, 3):
            plt.plot([i, i + 1], [price_tree[j, i], price_tree[j, i + 1]], 'k-')  # Up movements
            plt.plot([i, i + 1], [price_tree[j, i], price_tree[j + 1, i + 1]], 'k-')  # Down movements
            
            # Annotating the price at each node
        plt.text(i, price_tree[j, i], f'{price_tree[j, i]:.2f}', ha='center', va='bottom', fontsize=8)
        # Generate the binomial tree for a put option


plt.plot(0, S0, 'ro', label=f"Option Value: {option_value:.2f}")
plt.title('Binomial Tree for NSRGY Put Option Pricing')
plt.xlabel('Time Step')
plt.ylabel('Stock Price ($)')
plt.grid(True)
plt.show()

option_value
```

# Monte-Carlo Simulation

```python
# Parameters for the Monte Carlo simulation
S0 = spot_price_jan_2024  # Spot price from NSRGY data
K = strike_price  # Strike price calculated previously
T = time_to_maturity  # Time to maturity (assumed to be 1 month)
r = interest_rate  # Interest rate (3% annually)
sigma =log_volitality  # Volatility from NSRGY data
num_simulations = 10000  # Number of simulations

# Function to perform the Monte Carlo simulation for option pricing
def monte_carlo_option_pricing(S0, K, T, r, sigma, num_simulations):
    # Calculate the time step
    dt = T
    
    # Simulating random paths for the underlying asset price
    Z = np.random.standard_normal(num_simulations)
    ST = S0 * np.exp((r - 0.5 * sigma ** 2) * dt + sigma * np.sqrt(dt) * Z)
    
    # Calculating the payoff for a European put option at maturity
    payoff = np.maximum(K - ST, 0)
    
    # Discounting the payoff back to present value
    discount_factor = np.exp(-r * T)
    option_price = np.exp(-r * T) * np.mean(payoff)
    
    return option_price

# Running the simulation to get the option price
option_price = monte_carlo_option_pricing(S0, K, T, r, sigma, num_simulations)
print(f"The estimated European put option price using Monte Carlo simulation is: {option_price}")
```

# Black -scholes Formula

```python
# Define the Black-Scholes formula for a European put option
from scipy.stats import norm

def black_scholes_put(S0, K, T, r, sigma):
    """
    S0: current stock price
    K: strike price
    T: time to maturity
    r: risk-free interest rate
    sigma: volatility of the underlying stock
    """
    # Calculate d1 and d2
    d1 = (np.log(S0 / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    
    # Calculate the Black-Scholes put option price
    put_price = K * np.exp(-r * T) * norm.cdf(-d2) - S0 * norm.cdf(-d1)
    
    return put_price

# Given NSRGY data parameters
S0 = 113.90  # Spot price
K = 125.29  # Strike price
T = 1/12    # Time to maturity (1 month)
r = 0.03    # Risk-free interest rate
sigma = 0.1726  # Volatility

# Calculate the Black-Scholes put option price
black_scholes_put_price = black_scholes_put(S0, K, T, r, sigma)
black_scholes_put_price
```

# 1.DELTA

```python
# Parameters for the Black-Scholes formula
S0 = 113.90  # Current stock price
K = 125.29  # Strike price
T = 1/12  # Time to maturity (1 month)
r = 0.03  # Risk-free interest rate
sigma = 0.1726  # Volatility of the underlying stock

# Black-Scholes formula for calculating Delta
def black_scholes_delta(S0, K, T, r, sigma, option_type='call'):
    """
    S0: current stock price
    K: strike price
    T: time to maturity
    r: risk-free interest rate
    sigma: volatility of the underlying stock
    option_type: 'call' for call option delta, 'put' for put option delta
    """
    d1 = (np.log(S0 / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    if option_type == 'call':
        delta = norm.cdf(d1)
    elif option_type == 'put':
        delta = -norm.cdf(-d1)
    return delta

# Calculate the delta for a call and a put option
delta_call = black_scholes_delta(S0, K, T, r, sigma, option_type='call')
delta_put = black_scholes_delta(S0, K, T, r, sigma, option_type='put')

delta_call, delta_put
```

# 2.GAMMA

```python
# Black-Scholes Gamma formula implementation
def black_scholes_gamma(S0, K, T, r, sigma):
    """
    S0: current stock price
    K: strike price
    T: time to maturity
    r: risk-free interest rate
    sigma: volatility of the underlying stock
    """
    d1 = (np.log(S0 / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))
    gamma = norm.pdf(d1) / (S0 * sigma * np.sqrt(T))
    return gamma

# Parameters for Gamma calculation
S0 = 113.90  # Current stock price
K = 125.29  # Strike price
T = 1/12  # Time to maturity (1 month)
r = 0.03  # Risk-free interest rate
sigma = 0.1726  # Volatility

# Calculate Gamma
gamma = black_scholes_gamma(S0, K, T, r, sigma)

# Plotting Gamma against different stock prices
S_prices = np.linspace(80, 150, 100)  # Vector of stock prices
gamma_values = [black_scholes_gamma(S, K, T, r, sigma) for S in S_prices]

# 2D Plot for Gamma
plt.figure(figsize=(10, 5))
plt.plot(S_prices, gamma_values, label='Gamma')
plt.title('Gamma vs. Stock Price')
plt.xlabel('Stock Price')
plt.ylabel('Gamma')
plt.legend()
plt.grid(True)
plt.show()

# 3D Surface plot for Gamma across stock prices and time to maturity
T_maturities = np.linspace(1/365, 1, 100)  # Vector of times to maturity
S_grid, T_grid = np.meshgrid(S_prices, T_maturities)
gamma_surface = np.array([[black_scholes_gamma(S, K, T, r, sigma) 
                           for S, T in zip(S_row, T_row)] 
                          for S_row, T_row in zip(S_grid, T_grid)])

# 3D Plot for Gamma
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(S_grid, T_grid, gamma_surface, cmap='viridis')
ax.set_title('Gamma - 3D Surface Plot')
ax.set_xlabel('Stock Price')
ax.set_ylabel('Time to Maturity')
ax.set_zlabel('Gamma')
fig.colorbar(surf, shrink=0.5, aspect=5)  # Add a color bar which maps values to colors
plt.show()

gamma
```

# 3.Theta

```python
# First, let's implement the Theta calculation according to the Black-Scholes formula
def black_scholes_theta(S0, K, T, r, sigma, option_type='call'):
    """
    S0: current stock price
    K: strike price
    T: time to maturity
    r: risk-free interest rate
    sigma: volatility of the underlying stock
    option_type: 'call' for call option theta, 'put' for put option theta
    """
    d1 = (np.log(S0 / K) + (r + 0.5 * sigma ** 2) * T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    
    if option_type == 'call':
        theta = - (S0 * sigma * norm.pdf(d1)) / (2 * np.sqrt(T)) - r * K * np.exp(-r * T) * norm.cdf(d2)
    elif option_type == 'put':
        theta = - (S0 * sigma * norm.pdf(d1)) / (2 * np.sqrt(T)) + r * K * np.exp(-r * T) * norm.cdf(-d2)
    
    return theta / 365  # Convert to daily theta

# Parameters
S0 = 113.90  # Current stock price
K = 125.29  # Strike price
T = 1/12    # Time to maturity (1 month)
r = 0.03    # Risk-free interest rate
sigma = 0.1726  # Volatility

# Calculate Theta for both call and put options
theta_call = black_scholes_theta(S0, K, T, r, sigma, option_type='call')
theta_put = black_scholes_theta(S0, K, T, r, sigma, option_type='put')

# Now, let's plot Theta as a function of stock price and time to maturity
# Setting up a range for stock prices and time to maturity
S_prices = np.linspace(80, 150, 100)
T_maturities = np.linspace(1/365, 1, 100)

# 2D Plot for Theta vs. Stock Price at a fixed time to maturity
theta_call_values = [black_scholes_theta(S, K, T, r, sigma, 'call') for S in S_prices]
theta_put_values = [black_scholes_theta(S, K, T, r, sigma, 'put') for S in S_prices]

plt.figure(figsize=(10, 5))
plt.plot(S_prices, theta_call_values, label='Call Theta')
plt.plot(S_prices, theta_put_values, label='Put Theta')
plt.title('Theta vs. Stock Price')
plt.xlabel('Stock Price')
plt.ylabel('Theta')
plt.legend()
plt.grid(True)
plt.show()

# 3D Surface Plot for Theta
S_grid, T_grid = np.meshgrid(S_prices, T_maturities)
Theta_call_surface = np.array([[black_scholes_theta(S, K, T, r, sigma, 'call') for S in S_prices] for T in T_maturities])
Theta_put_surface = np.array([[black_scholes_theta(S, K, T, r, sigma, 'put') for S in S_prices] for T in T_maturities])

# Plot for Call Option Theta
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(S_grid, T_grid, Theta_call_surface, cmap='viridis')
ax.set_title('Call Option Theta - 3D Surface Plot')
ax.set_xlabel('Stock Price')
ax.set_ylabel('Time to Maturity')
ax.set_zlabel('Theta')
plt.show()

# Plot for Put Option Theta
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(S_grid, T_grid, Theta_put_surface, cmap='viridis')
ax.set_title('Put Option Theta - 3D Surface Plot')
ax.set_xlabel('Stock Price')
ax.set_ylabel('Time to Maturity')
ax.set_zlabel('Theta')
plt.show()
Theta_put_surface
Theta_call_surface
```
